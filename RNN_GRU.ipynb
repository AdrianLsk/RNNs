```python
import numpy as np
import numpy.random as rng

import theano
import theano.tensor as T

from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sb
sb.set()

import pandas as pd
import timeit
from pyeeg import embed_seq
```

```python
sunspots = sm.datasets.sunspots.load_pandas().data.astype('float32')
scaler = StandardScaler()
ts_train = scaler.fit_transform(sunspots.SUNACTIVITY)
x_train = embed_seq(ts_train, 1, 11)

rng = np.random
dtype = theano.config.floatX
n_samples, nn_input_dim = x_train.shape
nn_output_dim = nn_input_dim
n_h_dim = n_c_dim = 5

# Initialize Theano symbolic variables for the LSTM memory cell
inputs = T.matrix(dtype=dtype)
targets = T.matrix(dtype=dtype)
num_epochs = T.iscalar()
```

```python
# weights generator
def ortho_weights(sizeX, sizeY):
    # divide by larges singular value to account for vanishing gradients
    W = rng.uniform(low=-1., high=1.,  size=(sizeX, sizeY)).astype(dtype)
    s = np.linalg.svd(W)[1]
    W /= s[0]
    return W.astype(dtype)
```

```python
# parameters for in, out & forget gates:
Wrx = theano.shared(ortho_weights(n_h_dim, nn_input_dim), name='Wrx')  # stacked input-to-hidden unit weights
Wzx = theano.shared(ortho_weights(n_h_dim, nn_input_dim), name='Wzx')  # stacked input-to-hidden unit weights
Whx = theano.shared(ortho_weights(n_h_dim, nn_input_dim), name='Whx')  # stacked input-to-hidden unit weights
# hidden-to-hidden
Wrh = theano.shared(ortho_weights(n_h_dim, n_h_dim), name='Wrh')      # stacked hidden-to-hidden unit weights
Wzh = theano.shared(ortho_weights(n_h_dim, n_h_dim), name='Wzh')      # stacked hidden-to-hidden unit weights
Whh = theano.shared(ortho_weights(n_h_dim, n_h_dim), name='Whh')      # stacked hidden-to-hidden unit weights
# biases
br = theano.shared(np.zeros(n_h_dim,dtype=dtype), name='br') 
bz = theano.shared(np.zeros(n_h_dim,dtype=dtype), name='bz') 
bh = theano.shared(np.zeros(n_h_dim, dtype=dtype), name='bh')
# parameters for autoeconder ouput
Wyh = theano.shared(ortho_weights(nn_output_dim, n_h_dim), name='Wyh')  # stacked input-to-hidden unit weights
by = theano.shared(np.zeros(nn_output_dim, dtype=dtype), name='by')   

# initial values for h and c
h0 = theano.shared(value=np.zeros(n_h_dim, dtype=dtype)) # vector of initial values for memory cells

# deep learning LSTM tutorial, when used directly in scan's outputs_info for h0, c0
# [tensor.alloc(0., n_samples, n_h_dim), tensor.alloc(0., n_samples, n_h_dim)]

pars = [Wrx, Wzx, Whx, Wrh, Wzh, Whh, Wyh, br, bz, bh, by]

def gru_step(x_t, h_tm1, Wrx, Wzx, Whx, Wrh, Wzh, Whh, Wyh, br, bz, bh, by):
    # GRU cell
    pre_r = Wrx.dot(x_t) + Wrh.dot(h_tm1) + br
    pre_z = Wzx.dot(x_t) + Wzh.dot(h_tm1) + bz
    r = T.nnet.sigmoid(pre_r)
    z = T.nnet.sigmoid(pre_z)
    pre_h = Whx.dot(x_t) + Whh.dot(r * h_tm1) + bh
    h = z * h_tm1 + (1 - z) * T.tanh(pre_h)
    y = Wyh.dot(h) + by # RNN regression output
    # RNN classification output y = T.nnet.softmax(Wyh.dot(h) + by)[0]
    return [h, y]

[h_vals, y_vals] = theano.scan(fn=gru_step,
                              sequences=[dict(input=inputs, taps=[0])],
                              outputs_info=[h0, None],
                              non_sequences=[Wrx, Wzx, Whx, Wrh, Wzh, Whh, Wyh, br, bz, bh, by],
                              strict=True)[0]

#  + weight_regularization * (w ** 2).sum() + hidden_regularization * T.abs(w).sum() ADD all weights
cost = ((inputs - y_vals)**2).mean(axis=0).sum() # regression

# learning rate
lr = 0.1
learning_rate = theano.shared(np.cast[dtype](lr))

# gradients
gpars = []
for par in pars:
    gpar = T.grad(cost, par)
    gpars.append(gpar)

# pars updates
updates=[]
for par, gpar in zip(pars, gpars):
    updates.append((par, par - gpar * learning_rate))

# compile traning function
learn_gru_rnn = theano.function(inputs=[inputs],
                                 outputs=[h_vals, cost],
                                 updates=updates,
                                 allow_input_downcast=True)

# prediction function
forecast = Wyh.dot(h_vals.T).T+ by
predict_gru_rnn = theano.function(inputs=[inputs],
                                   outputs=forecast,
                                   allow_input_downcast=True
                                  )
```

```python
# visualize training epochs
plt.figure(0, figsize=[15,5])
plt.plot(x_train[:,-1])
n_epochs=1000
train_errors = []
for x in xrange(n_epochs):
    train_errors.append(np.asscalar(learn_gru_rnn(x_train)[1]))
    if x % 100:
        plt.plot(predict_gru_rnn(x_train)[:,-1], color=np.random.rand(3,1))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[15, 5], num=1)
ax1.plot(x_train[:,-1], 'r-', label='truth')
ax1.plot(predict_gru_rnn(x_train)[:,-1], label='lstm rnn')
ax1.legend(loc=2)
ax2.plot(np.arange(n_epochs), train_errors, 'g-')
ax2.set_xlabel('epochs')
ax2.set_ylabel('error')
ax2.set_ylim(0., 50)
ax2.legend(['error'], loc=2)
fig.show()
```
